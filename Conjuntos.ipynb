# Um olhar físico sobre Teoria de Conjuntos

## 1 &nbsp; Descrição
Imagine por um momento, que você é um cara muito atrelado a física e só consegue entender os conceitos utilizando formalizações quasi-reais para racionalizar bem sobre um problema. Então, como você explicaria Teoria de Conjuntos? Fácil, utilizaria conceitos de física!

## 2 &nbsp; Situando-se no problema 
A priori, se vamos pensar como físicos, vamos imaginar como poderiamos começar a modelar os conjuntos. Veja, veja, eu pensei da seguinte forma:



Conjunto geométrico.jpg


Pensar a noção de conjunto como orbitas seria uma boa, já que podemos atrelar a isso alguns paramêtros como gravidade e velocidade tangencial, e com isso fazer algumas relações matemáticas! Calma que já explico cada coisa.
Basicamente, cada circunfêrencia representa um conjunto e cada ponto nessa circunferência é um elemento, assim como cada arco representa um subconjunto.
